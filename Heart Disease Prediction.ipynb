```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the dataset
df = pd.read_csv("HeartDiseaseTrain-Test.csv")  # Update the filename if needed

# Display first few rows
df.head()
```

```python
print(df["sex"].unique())
print(df["chest_pain_type"].unique())
print(df["fasting_blood_sugar"].unique())
print(df["rest_ecg"].unique())
print(df["exercise_induced_angina"].unique())
print(df["slope"].unique())
print(df["vessels_colored_by_flourosopy"].unique())
print(df["thalassemia"].unique())
print(df["target"].unique())
```

```python
df["sex"]=df["sex"].map({"Male":1,"Female":0})
df["chest_pain_type"]=df["chest_pain_type"].map({"Typical angina":0,"Atypical angina":1,"Non-anginal pain":2,"Asymptomatic":3})
df["rest_ecg"]=df["rest_ecg"].map({"ST-T wave abnormality":0 ,"Normal":1,"Left ventricular hypertrophy" :2})
df["fasting_blood_sugar"]=df["fasting_blood_sugar"].map({"Lower than 120 mg/ml":0,"Greater than 120 mg/ml":1})
df["exercise_induced_angina"]=df["exercise_induced_angina"].map({'No':0, 'Yes':1})
df["slope"]=df["slope"].map({'Downsloping':0, 'Upsloping':1 ,'Flat':2})
df["vessels_colored_by_flourosopy"]=df["vessels_colored_by_flourosopy"].map({'Two':0 ,'Zero':1, 'One' :2,'Three':3,'Four':4
})
df["thalassemia"]=df["thalassemia"].map({'Reversable Defect' :0,'Fixed Defect':1, 'Normal':2, 'No':3})




```

```python
df.head()
```

```python
#pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
x = df.drop(columns=['target'])  # Assuming 'target' is the column for heart disease
y = df['target']
# Split data into training and testing sets
# Split data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Check shapes
print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Create Decision Tree model
model = DecisionTreeClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

print("Model training complete!")
```

```python
from sklearn.metrics import accuracy_score

# Make predictions on test set
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy*100:.2f}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Disease", "Disease"], yticklabels=["No Disease", "Disease"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(15, 8))

# Plot the decision tree
plot_tree(model, feature_names=x.columns, class_names=["No Disease", "Disease"], filled=True)

plt.show()
```

```python
def predict_heart_disease(model):
    print("\n Enter your health details (Use numbers as instructed):")

    # Numeric Inputs
    age = int(input("Age: "))
    resting_blood_pressure = int(input("Resting Blood Pressure: "))
    cholestoral = int(input("Cholesterol Level: "))
    Max_heart_rate = int(input("Max Heart Rate Achieved: "))
    oldpeak = float(input("ST Depression Induced by Exercise: "))

    # Categorical Inputs (User enters numbers)
    sex = int(input("Sex (Male: 1, Female: 0): "))
    chest_pain_type = int(input("Chest Pain Type (Typical Angina: 0, Atypical Angina: 1, Non-Anginal Pain: 2, Asymptomatic: 3): "))
    fasting_blood_sugar = int(input("Fasting Blood Sugar (Lower than 120 mg/ml: 0, Greater than 120 mg/ml: 1): "))
    rest_ecg = int(input("Resting ECG (ST-T wave abnormality: 0, Normal: 1, Left ventricular hypertrophy: 2): "))
    exercise_induced_angina = int(input("Exercise-Induced Angina (No: 0, Yes: 1): "))
    slope = int(input("Slope of Peak Exercise ST Segment (Downsloping: 0, Upsloping: 1, Flat: 2): "))
    vessels_colored_by_flourosopy = int(input("Number of Major Vessels (Zero: 0, One: 1, Two: 2, Three: 3, Four: 4): "))
    thalassemia = int(input("Thalassemia Type (Reversible Defect: 0, Fixed Defect: 1, Normal: 2, No: 3): "))

    # Create a DataFrame with the users input
    import pandas as pd
    user_data = [[age, sex, chest_pain_type, resting_blood_pressure, cholestoral, fasting_blood_sugar,
                  rest_ecg, Max_heart_rate, exercise_induced_angina, oldpeak, slope, vessels_colored_by_flourosopy, thalassemia]]

    user_df = pd.DataFrame(user_data, columns=x.columns)  # Ensure column names match

    # Make prediction
    prediction = model.predict(user_df)
    print("\nPredicted Heart Disease: ", prediction)

    # Output result
    if prediction[0] == 1:
        print("\n High risk of heart disease! Please consult a doctor.")
    else:
        print("\n No signs of heart disease detected. Stay healthy!")
```

```python
predict_heart_disease(model)
```
